# American Community Survey 2022 Microdata Analysis Report
> This notebook contains the summary report of findings from the analysis of the 2022 American Community Survey Microdata Sample.

* Stakeholder questions that is analysis aims to answer include:
    * Which states have the greatest sample population size and what percent of the sample do they take up?
    * How does race, gender, and disability impact education attainment and employment?
    * Is there a relationship between education and income?
    * What is the usual age group per each level of attained education?
    * Is there a relationship between household income and the costs that each household pays?
    * Is there a correlation between education attainment/enrollment and accesss to technology?

> Each section of this notebook will present findings based on each question asked above.

## Education Attainment and Employment by Sex, Race, and Disability Status

For this part of the analysis, since we're analyzing employment percentages, we'll use a subset of the sample poplation to only include individuals aged *25-54*, which is the prime employment age range according the Bureau of Labor Statistics.

Also for this analysis, _unemployment_ will be categorized as someone who:
* NILF who last worked more than 5 years ago or never worked
    * NILF = **N**ot **I**n **L**abor **F**orce
* Unemployed and last worked 5 years ago or earlier or never worked
* Working without pay in family business or farm

We will use the prime employment age range (25-54) for this part of the analysis as well, since we are analyzing employment.

## Relationship between Household Income and Household Costs

For this section, the relationships between income and household costs were calculated by using the Pearson Correlation Coefficient formula:
![alt text](../images/pearson_correlation_coefficient_formula.png)
Where income was always x and the household incomes were always y.


| x      | y                        |r          |
|:-------|:-------------------------|:----------|
| income | monthly rent             | 0.48326   |
| income | monthly electricity cost | 0.12667   |
| income | monthly gas cost         | 0.0543069 |
| income | annual water cost        | 0.14671   |
| income | property taxes           | 0.447747  |

From these correlation coefficients, we can infer that there is a mild relationship between income and monthly rent, and income and property taxes. The relationship between income and the other cost values are very weak or almost nonexistent.

## Relationship between Access to Technology and Higher Education

> This section of the analysis will examine how access to technologies, such as computers or internet, may impact the ability to attain higher education.

> The analysis will be conducted on a state by state basis.

> Since this section will not focus on impact on employment, but higher education, the age range of individuals will be 18 and up.

> Also for this section, "Attained Higher Education" will be considered as having at least an Associate's Degree.

> And "Enrolled in Higher Education" will be considered as being enrolled in college courses, even if the individual has not yet attained an Associate's Degree.

### Computer Access and Education
![alt text](../images/comp_access_edu.png)
* Higher Education Attainment:
    * In most states, the trend is as percent of households that have a computer increases, the percent of attained higher education increases.
    * The visualization and the correlation coefficient of approximately 0.82 suggests a strong positive relationship between computer access and higher education attainment.

* Higher Education Enrollment:
    * In most states, the trend is as percent of households that has a computer increases, the percent of those enrolled in higher education _slightly_ increases.
    * The visualization and the correlation coefficient of approximately 0.518 suggests a mild positive relationship between computer access and higher education enrollment.

### Smartphone Access and Education

* Higher Education Enrollment:
    * The visualization does not depict a clear trend between percent of households with smartphone access and higher education enrollment by state.
